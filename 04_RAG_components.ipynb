### Vector Store 
Several databases are now available such as Chroma, Postgres, etc.
- They store the embeddingsin a very fast searchabble format.
- Install package for *langcahin-chroma == 0.1.2* 

```python
# from langchain.vectorstores import Chroma -outdated-version
from langchain_chroma import Chroma
from langchain_openai import OpenAIEmbeddings
from langchain_text_splitters import CharacterTextSplitter
from langchain_chroma import Chroma

#Load the documents, split it into chunks, embed each chunk and load data into vector store
loaded_document = TextLoader("./data/state_of_the_union.txt").load()

# Create splitter
text_splitter = CharacterTextSplitter(chunk_size=1000, chunk_overlap=0)

# create chunks for each document using the splitter. 
chunks_of_text = text_splitter.split_documents(loaded_document)

```

```python
len(chunks_of_text)
```

```python

# craete embeddings with embeddings Model and load into the database to 
# form a *Vector Store*
vector_db = Chroma.from_documents(chunks_of_text, OpenAIEmbeddings())
```

```python
# Vector db object
vector_db
```

#### Performing the Similarity Search based on queries

```python
question = "What did the president say about the John Lewis Voting Rights Act"

response = vector_db.similarity_search(question)

# print response
print(response[0].page_content)
```

### Vector Store as Retriver
* Find the embedding that best answers your question

```python
from langchain_community.document_loaders import TextLoader

loader = TextLoader("./data/state_of_the_union.txt")
```

### Using FAIS Vector DB

```python
from langchain_community.vectorstores import FAISS  #install FAISS package with *pip*
from langchain_openai import OpenAIEmbeddings
from langchain_text_splitters import CharacterTextSplitter

loaded_doocument = loader.load()

# Create splitter
text_splitter = CharacterTextSplitter(chunk_size=1000, chunk_overlap=0)

# create chunks for each document using the splitter. 
chunks_of_text = text_splitter.split_documents(loaded_document)

emmbeddings = OpenAIEmbeddings()

vector_db = FAISS.from_documents(chunks_of_text,emmbeddings)
```

```python
# Create retriever *as_retriever* function
# used everytime we perform a RAG operation
# more powerful than similarity search function.
retriever = vector_db.as_retriever(search_kwargs={"k":2})   #Top K = 2
```

```python
response = retriever.invoke("What did he say about Ketanji Brown Jackson?")
response
```
